```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#import seaborn as sns
%matplotlib inline
#sns.set_style('white')
```

```python
url = 'http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv'
data = pd.read_csv(url)
data.head()
```

```python
stem = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3), facecolor='white')

rgb_dark_blue = (0/255, 107/255, 164/255)
rgb_orange = (255/255, 128/255, 14/255)

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(data['Year'], data[stem[sp]], c= rgb_orange, label='Women', linewidth=3)
    ax.plot(data['Year'], (100-data[stem[sp]]), c= rgb_dark_blue , label='Men', linewidth=3)
    for key,spine in ax.spines.items():
        spine.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    #ax.set_axis_bgcolor('white')
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')

#plt.legend(loc='upper right')
plt.show()
```
