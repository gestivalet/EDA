# Unemployment Rate during the Obama Administration

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#value = in percent
sns.set_style('white')
```

```python
data = pd.read_csv('~/desktop/Python/Data Quest/unrate.csv')
data['DATE'] = pd.to_datetime(data['DATE'])
data.head(3)
```

```python
obama_adm = data[data['DATE'] >= '2009-01-01']
obama_adm = obama_adm.reset_index().drop('index',axis=1)
print obama_adm.head(12)
print obama_adm.tail(12)
```

It's already possible to see that the unemployment rate has fallen since the beginning of the Obama Administration.

```python
fig, ax = plt.subplots(figsize=(18,5))
ax.plot(obama_adm.DATE, obama_adm.VALUE, linewidth=4)
ax.set_title('Unemployment Rate Over Obama Administration', fontsize=15)
ax.set_xlabel('Years', fontsize=13); ax.set_ylabel('Unemployment Rate in %', fontsize=13)
ax.text('2009', '7.3', '7.8%', fontsize=15)
ax.text('2017', '4.8', '4.7%', fontsize=15)

for key,spine in ax.spines.items():
        spine.set_visible(False)

plt.show()
```
