```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
url = 'https://kaggle2.blob.core.windows.net/competitions-data/kaggle/3136/train.csv?sv=2015-12-11&sr=b&sig=piIZkGUpxE8pmu74xS9FkkIBzJa8I9YSFPl6gf%2FZzrs%3D&se=2017-01-27T15%3A13%3A04Z&sp=r'
titanic = pd.read_csv(url)
titanic = titanic.drop(['PassengerId','Name','Cabin','Ticket'], axis=1)
titanic = titanic.dropna()
titanic.head(2)
```

```python
sns.distplot(titanic.Age)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
sns.kdeplot(titanic['Age'], shade=True)
sns.despine(left=True, bottom=True)
plt.xlabel('Age')
plt.show()
```

```python
g = sns.FacetGrid(titanic, col='Pclass', size=6)
g.map(sns.kdeplot, 'Age', shade=True)
sns.despine(left=True, bottom=True)
plt.show()
```

```python
g = sns.FacetGrid(titanic, row='Pclass',col="Survived", hue='Sex', size=4, )
g.map(sns.kdeplot, 'Age', shade=True).add_legend()
sns.despine(left=True, bottom=True)
plt.show()
```
