```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('white')
```

```python
url = 'https://raw.githubusercontent.com/fivethirtyeight/data/master/college-majors/recent-grads.csv'
data = pd.read_csv(url)
data.head()
```

```python
data.columns.value_counts().sort_index()
```

```python
data.describe()
```

```python
raw_data_count = data.shape[0]
print 'Before: ',raw_data_count
data = data.dropna()
cleaned_data_count = data.shape[0]
print 'Cleaned: ',cleaned_data_count
```

```python
data.describe()
```

```python
data.plot(x='Sample_size', y='Median', kind='scatter')
```

```python
fig, ( (ax1, ax2, ax3), (ax4, ax5, ax6) ) = plt.subplots(2,3, figsize=(18,6))

ax1.scatter(x=data['Sample_size'], y=data['Median'])
ax1.set_xlabel('Sample Size'); ax1.set_ylabel('Median')
for key,spine in ax1.spines.items():
        spine.set_visible(False)

ax2.scatter(data['Sample_size'], data['Unemployment_rate'])
ax2.set_xlabel('Sample Size'); ax2.set_ylabel('Unemployment Rate')
for key,spine in ax2.spines.items():
        spine.set_visible(False)

ax3.scatter(data['Full_time'], data['Median'])
ax3.set_xlabel('Full Time'); ax3.set_ylabel('Median')
for key,spine in ax3.spines.items():
        spine.set_visible(False)

ax4.scatter(data['ShareWomen'], data['Unemployment_rate'])
ax4.set_xlabel('Share of Women'); ax4.set_ylabel('Unemployment Rate')
for key,spine in ax4.spines.items():
        spine.set_visible(False)

ax5.scatter(data['Men'], data['Median'])
ax5.set_xlabel('Men'); ax5.set_ylabel('Median')
for key,spine in ax5.spines.items():
        spine.set_visible(False)

ax6.scatter(data['Women'], data['Median'])
ax6.set_xlabel('Women'); ax6.set_ylabel('Median')
for key,spine in ax6.spines.items():
        spine.set_visible(False)

plt.show()
```

Use the plots to explore the following questions:
Do students in more popular majors make more money?
Do students that majored in subjects that were majority female make more money?
Is there any link between the number of full-time employees and median salary?

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]

fig = plt.figure(figsize=(18,16))
for r in range(8):
    ax = fig.add_subplot(4,2,r+1)
    ax = data[cols[r]].plot(kind='hist', rot=0, bins=20)
    ax.set_title(cols[r])
    for key,spine in ax.spines.items():
        spine.set_visible(False)
        
plt.show()
```
